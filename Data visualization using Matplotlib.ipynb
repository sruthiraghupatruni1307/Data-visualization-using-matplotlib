```python
import pandas as pd
url="https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Stock/MSFT_data.csv"
```

```python
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(25,10))
df1=df.head(60)
plt.plot(df1['date'],df1['open'])
plt.show()
```

```python
plt.figure(figsize=(25,10))
df1=df.head(60)
plt.plot(df1['date'],df1['open'])
plt.xlabel('date')
plt.ylabel('opening price of MSFT stock')
plt.title('date vs open price')
plt.show()
```

```python
plt.figure(figsize=(25,10))
df1=df.head(60)
plt.plot(df1['date'],df1['open'])
plt.xlabel('date')
plt.ylabel('opening price of MSFT stock')
plt.title('date vs open price')
plt.xticks(df1['date'],rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(25,10))
df1=df.head(60)
df2=df.tail(60)
plt.plot(df1['date'],df1['open'])
plt.plot(df2['date'],df2['open'])
plt.xlabel('date')
plt.ylabel('opening price of MSFT stock')
plt.title('date vs open price')
plt.xticks(df1['date'],rotation='vertical')
plt.show()
```

```python
x=np.arange(-55,55)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(x)")
```

```python
help(plt.plot)
```

```python
x=np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>')
plt.plot(x,x**3,c='blue',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(x)")
```

```python
x=np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>',linestyle='-.')
plt.plot(x,x**3,c='blue',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(x)")
```

```python
x=np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>',linestyle='-.')
plt.plot(x,x**3,c='blue',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(x)")
plt.legend(['Y=X**2','Y=X**3'])
plt.show()
```

```python
x=np.arange(-5,10)
plt.plot(x,x**2,c='black',marker='>',linestyle='-.')
plt.plot(x,x**3,c='blue',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(x)")
plt.legend(['Y=X**2','Y=X**3'],loc='upper center')
plt.text(1,1,"(1,1)")
plt.text(5,25,"(5,25)")
plt.grid() # plt.axis("off")
plt.show()
```

## Scatter plot
identifing the relationship between two features
- +ve relationship
- -ve relationship
- neutral

### vectorization

```python
x=[1,2,3,4,5]
y=np.array([2,4,6,8,10])
y2=-y
y3=[5,5,5,5,5]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y2)
plt.show()
```

```python
plt.scatter(x,y3)
plt.show()
```

```python
plt.scatter(x,y,c=['b','r','g','black','yellow'],alpha=0.9)
```

```python
plt.scatter(x,y,marker='>',c=['b','r','g','purple','yellow'],alpha=0.9,s=y**3,edgecolors='black')
plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.scatter(x,y,marker='>',c=['b','r','g','purple','yellow'],alpha=0.9,s=y**3,edgecolors='black')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x,y,marker='>',c=['b','r','g','purple','yellow'],alpha=1.0,s=y**3,edgecolors='black')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(x)")
plt.legend(['Y=X**2'],loc='upper center')
plt.text(1,1,"(1,1)")
plt.text(5,2,"(5,25)")
plt.grid(color='#aa0000',linestyle='-',linewidth=2) # plt.axis("off")
plt.show()
```
